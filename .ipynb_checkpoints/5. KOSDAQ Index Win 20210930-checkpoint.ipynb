- #### Naver Price Data

```python
# Get the data from Nave Chart
def make_price_data(code, timeframe, count):
    url = 'https://fchart.stock.naver.com/sise.nhn?symbol=' + code + '&timeframe=' + timeframe + '&count=' + count + '&requestType=0'
    price_data = requests.get(url)
    price_data_bs = bs4.BeautifulSoup(price_data.text, 'lxml')
    item_list = price_data_bs.find_all('item')

    date_list = []
    open_list = []
    high_list = []
    low_list = []
    close_list = []
    trade_list = []

    for item in item_list:
        data = item['data'].split('|')
        date_list.append(data[0])
        open_list.append(data[1])
        high_list.append(data[2])
        low_list.append(data[3])
        close_list.append(data[4])
        trade_list.append(data[5])        

    price_df = pd.DataFrame({'open': open_list, 'high': high_list, 'low': low_list, 'close': close_list, 'volume': trade_list}, index=date_list)            
    price_df['code'] = code
    num_vars = ['open','high','low','close','volume']
    char_vars = ['code']
    price_df = price_df.reindex(columns = char_vars + num_vars)

    for var in num_vars:
        price_df[var] = pd.to_numeric(price_df[var], errors='coerce')

    price_df.index = pd.to_datetime(price_df.index, errors='coerce')

    return price_df

kosdaq_list = pd.read_pickle('kosdaq_code.pkl')

price_data = pd.DataFrame()
for code in kosdaq_list['code']:  
    daily_price = make_price_data(code, 'day', '300')
    price_data = pd.concat([price_data, daily_price], axis=0)   

price_data.index.name = 'date'
price_data.to_pickle('price_data.pkl')
```

- #### Merge Stock Price Data & KOSPI index Data

```python
kosdaq_list = pd.read_pickle('kosdaq_code.pkl')

price_kosdaq = price_data.merge(kosdaq_index['kosdaq_return'], left_index=True, right_index=True, how='left') # merge individual stock price data with KOSPI index

# flag whether daily price return is better than market price return
return_all = pd.DataFrame()
for code in kosdaq_list['code']:  
    stock_return = price_kosdaq[price_kosdaq['code']==code].sort_index()
    stock_return['return'] = stock_return['close']/stock_return['close'].shift(1)
    c1 = (stock_return['kosdaq_return'] < 1)
    c2 = (stock_return['return'] > 1)
    stock_return['win_market'] = np.where((c1&c2), 1, 0)
    return_all = pd.concat([return_all, stock_return], axis=0) 
    
return_all.to_pickle('return_all.pkl')    
```

```python
kosdaq_list = pd.read_pickle('kosdaq_code.pkl')

# with open('rf', 'rb') as file:  
#     rf = pickle.load(file)   

def strategy_simul(start_date, end_date):

    w_days = list(pd.bdate_range(start_date, end_date)) #  

    results_all = pd.DataFrame()

    for dt in w_days: #      


        obs_end = datetime.datetime.strftime(dt, "%Y-%m-%d")
        start_date = datetime.datetime.strptime(obs_end, "%Y-%m-%d") - datetime.timedelta(days=180) #  180  
        obs_start = datetime.datetime.strftime(start_date, "%Y-%m-%d")      

        code_list = []
        
        price_list = []
        price_mean_list = []
        price_std_list = []
        price_z_list = []
        price_z_out_list = []
        price_5ma_list = []
        price_low_list = []
        price_high_list = []
        
        volume_list = []
        volume_mean_list = []
        volume_std_list = []
        volume_z_list = []
        volume_z_out_list = []
        
        wins_60_list = []
        wins_180_list = []

        toptail_list = []
        
        for code in kosdaq_list['code']:
            s = return_all[return_all['code']==code].loc[obs_start:obs_end]

            if (s['volume'].sum() < 1000000) or (s['close'].std() == 0) or (s['low'].min() == 0): # 6     ...
                continue

            code_list.append(code)

            price_list.append(s['close'].tail(1).max()) #  
            price_mean_list.append(s['close'].mean()) #  
            price_std_list.append(s['close'].std()) #  
            price_z =  (s['close'].tail(1).max() - s['close'].mean()) / s['close'].std() #  Z 
            price_z_out =  np.where( ((s['close'] - s['close'].mean()) / s['close'].std()) > 1.6, 1, 0) #    
            price_5ma_list.append(s['close'].tail(5).mean()) # 5 
            price_low_list.append(s['low'].tail(1).max()) #     
            price_high_list.append(s['high'].tail(1).max()) #     
    
            volume_list.append(s['volume'].tail(1).max()) #  
            volume_mean_list.append(s['volume'].mean()) #  
            volume_std_list.append(s['volume'].std()) #  
            volume_z = (s['volume'].tail(1).max() - s['volume'].mean()) / s['volume'].std()  #  Z 
            volume_z_out = np.where( ((s['volume'] - s['volume'].mean()) / s['volume'].std()) > 1.6, 1, 0) #     

            price_z_list.append(price_z)
            volume_z_list.append(volume_z)

            price_z_out_list.append(price_z_out.sum())
            volume_z_out_list.append(volume_z_out.sum())

            wins_60_list.append(s['win_market'].tail(60).sum()) #  180       
            wins_180_list.append(s['win_market'].sum()) #  60       
            
            toptail = (s['close'] > s['open'])*((s['high']/s['close'])>1.05).astype('int') #  5%    
            toptail_list.append(toptail.sum())
            

        #   -  
        wins = pd.DataFrame({'code': code_list, 'price': price_list, 'price_mean': price_mean_list, 'price_std': price_std_list, 'price_z': price_z_list,'price_z_out': price_z_out_list,\
                             'price_5ma': price_5ma_list, 'price_high': price_high_list, 'price_low': price_low_list,\
                             'volume': volume_list, 'volume_mean': volume_mean_list, 'volume_std': volume_std_list, 'volume_z': volume_z_list, 'volume_z_out': volume_z_out_list,\
                             'num_wins_60': wins_60_list, 'num_wins_180': wins_180_list, 'num_toptail': toptail_list})


        wins['buy_price'] = wins['price']*0.985 #     1.5%  

        c1 = (wins['num_wins_180'] > 20)  #  180        > 20 ( )
        c2 = (wins['num_wins_60'] > 10)  #  180        > 20 ( )
        candidates = wins[ c1 & c2 ] #   
        

        out_start =  datetime.datetime.strftime(datetime.datetime.strptime(obs_end, "%Y-%m-%d") + datetime.timedelta(days=1), "%Y-%m-%d") # 
        out_end  =  datetime.datetime.strftime(datetime.datetime.strptime(obs_end, "%Y-%m-%d") + datetime.timedelta(days=7), "%Y-%m-%d") #  

        profit_sell = 1.08 # 
        loss_sell = 0.96 # 

        outcome_all = pd.DataFrame()   
       
        for index, row in candidates.iterrows():

            outcome = return_all[return_all['code']==row['code']].loc[out_start : out_end][['code', 'low', 'high', 'close']] #  
            outcome['buy_price'] = int(row['buy_price']) #  
            outcome['highest'] = outcome['high'].max()
            outcome['lowest'] = outcome['low'].min()
            outcome['price_z'] = row['price_z'] #  
            outcome['volume_z'] = row['volume_z'] #  
            outcome['price_z_out'] = row['price_z_out'] # 180  
            outcome['price_5ma'] = row['price_5ma'] # 5  
            outcome['volume_z_out'] = row['volume_z_out'] # 180  
            outcome['num_wins_60'] = row['num_wins_60'] # 60  
            outcome['num_wins_180'] = row['num_wins_180'] # 180       
            outcome['num_toptail'] = row['num_toptail'] # 180       
            outcome['num_wins_trend'] = outcome['num_wins_60'] / outcome['num_wins_180']
            outcome['price_from_5ma'] = outcome['buy_price'] / outcome['price_5ma']
            outcome['price_high'] = row['price_high'] #  
            outcome['price_low'] = row['price_low'] #  
            outcome['price_change'] = outcome['price_high']/outcome['price_low']  #            
            outcome['yymmdd'] = obs_end

            
            indep = ['price_change','price_z', 'volume_z', 'price_z_out', 'volume_z_out', 'num_wins_60', 'num_wins_trend', 'price_from_5ma','num_toptail']
            
#             for var in indep:
#                 bins =  [ -np.inf, bins_dict[var]['q1'], bins_dict[var]['q2'], bins_dict[var]['q3'], bins_dict[var]['q4'], np.inf]
#                 outcome['x_'+ var] = pd.cut(outcome[var], bins = bins, labels=range(5))
                                            
#             cols = [c for c in outcome.columns if 'x_' in c]
#             X = outcome[cols]

#             outcome['yhat'] = rf.predict_proba(X)[:,1]
#             outcome['select'] = np.where(outcome['yhat']  > 0.5, 1, 0)
            
            outcome['buy'] = np.where(outcome['low'].head(1).max() < outcome['buy_price'].max(), 1, 0) #     ?
            
            outcome['return_close'] = outcome['close'] / outcome['buy_price'] #  : (  / )
            outcome['return_high'] = outcome['high'] / outcome['buy_price'] #  :  / )   
            outcome['return_low'] = outcome['low'] / outcome['buy_price'] #  :  / )                                                                                                                                                                                    
            
            outcome['profit_sell'] = np.where(outcome['return_high'] > profit_sell, 1, 0)  #    ?
            outcome['loss_sell'] = np.where(outcome['return_close'] < loss_sell, 1, 0)  #    ?     
                                                 
            outcome_all = pd.concat([outcome_all, outcome], axis=0)         
                                          
                  
        #   
        for code in list(outcome_all['code'].unique()):   
            code_df = outcome_all[outcome_all['code']==code].sort_index()
            profit = code_df[code_df['profit_sell']==1].head(1)  #   
            loss = code_df[code_df['loss_sell']==1].head(1) #   
            final = code_df.tail(1) #     

            results = pd.concat([profit, loss, final], axis=0).sort_index() #   
            results_all = pd.concat([results_all, results], axis=0) #   
            results_all = results_all.sort_index().drop_duplicates(subset=['yymmdd','code'], keep='first')  #                                      
 

    return outcome_all, results_all
                                            
# outcome_202105, results_202105 = strategy_simul('2021-05-01', '2021-05-31')
# results_202105.to_pickle('results_202105.pkl')

# outcome_202106, results_202106 = strategy_simul('2021-06-01', '2021-06-30')
# results_202106.to_pickle('results_202106.pkl')

# outcome_202107, results_202107 = strategy_simul('2021-07-01', '2021-07-31')
# results_202107.to_pickle('results_202107.pkl')

# outcome_202108, results_202108 = strategy_simul('2021-08-01', '2021-08-31')
# results_202108.to_pickle('results_202108.pkl')

# outcome_202109, results_202109 = strategy_simul('2021-09-01', '2021-09-30')
# results_202109.to_pickle('results_202109.pkl') 

# outcome_202110, results_202110 = strategy_simul('2021-10-01', '2021-10-26')
# results_202110.to_pickle('results_202110.pkl') 
```

- ### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

pd.options.display.float_format = '{:,.3f}'.format

results_202105 = pd.read_pickle('results_202105.pkl')
results_202106 = pd.read_pickle('results_202106.pkl')
results_202107 = pd.read_pickle('results_202107.pkl')
results_202108 = pd.read_pickle('results_202108.pkl')
results_202109 = pd.read_pickle('results_202109.pkl')

profit_sell = 1.08 # 
loss_sell = 0.96 # 

df = pd.concat([results_202105, results_202106, results_202107, results_202108, results_202109], axis=0)
df['return'] = np.where(df['profit_sell']==1, profit_sell, np.where(df['loss_sell']==1, loss_sell, df['return_close']))
model_data = df[df['buy']==1]

print(model_data.groupby('yymmdd')['return'].mean().product())
print(model_data.groupby('yymmdd')['return'].mean().mean())


model_data['actual'] = np.where(model_data['return']>1, 1, 0)

indep = ['price_change','price_z','volume_z','price_z_out','volume_z_out','num_wins_60','num_wins_trend','price_from_5ma','num_toptail']

y = model_data['actual']
X = model_data[indep]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, shuffle=True, random_state=24)

rf = RandomForestClassifier(n_estimators=300, bootstrap = True, min_samples_leaf = 25, max_features = 5, max_depth = 5, min_samples_split = 50)

# Fit on training data
rf.fit(X_train, y_train)
print(rf.get_params())

yhat_train = pd.Series(rf.predict_proba(X_train)[:,1], index=y_train.index, name='yhat')
yhat_test = pd.Series(rf.predict_proba(X_test)[:,1], index=y_test.index, name='yhat')

comb_train = pd.concat([y_train, yhat_train], axis=1)
comb_test = pd.concat([y_test, yhat_test], axis=1)

print('Train # of Samples: ', len(comb_train))
print('Test # of Samples: ', len(comb_test))

comb_train['ranks'] = pd.qcut(comb_train['yhat'], q=10, duplicates='drop')
comb_test['ranks'] = pd.qcut(comb_test['yhat'], q=10, duplicates='drop')

print(comb_train.groupby('ranks')['actual'].mean())
print(comb_test.groupby('ranks')['actual'].mean())


# with open('rf', 'wb') as file:  
#     pickle.dump(rf, file)
    
# with open('rf', 'rb') as file:  
#     rf = pickle.load(file)   
```

```python
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True, random_state=24)


param_grid = {'bootstrap': [True], 'max_depth': [7, 8],  'max_features': [8, 9], 'min_samples_leaf': [30, 40], 'n_estimators': [200, 300],  'min_samples_split' : [50, 40]}

# Create a based model
rf = RandomForestClassifier()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 3, n_jobs = -1, verbose = 2)
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)

yhat_train = pd.Series(grid_search.predict_proba(X_train)[:,1], name='yhat', index=y_train.index)
comb_train = pd.concat([y_train, yhat_train], axis=1)

yhat_test = pd.Series(grid_search.predict_proba(X_test)[:,1], name='yhat', index=y_test.index)
comb_test = pd.concat([y_test, yhat_test], axis=1)

print('Train # of Samples: ', len(comb_train))
print('Test # of Samples: ', len(comb_test))

# Performance Test
comb_train['ranks'] = pd.qcut(comb_train['yhat'], q=10, duplicates='drop')
comb_test['ranks'] = pd.qcut(comb_test['yhat'], q=10, duplicates='drop')
print(comb_train.groupby('ranks')['actual'].mean())
print(comb_test.groupby('ranks')['actual'].mean())

with open('rf', 'wb') as file:  
    pickle.dump(grid_search, file)
    
with open('rf', 'rb') as file:  
    rf = pickle.load(file)   
```

```python
import timeit
from IPython.display import clear_output
import glob

from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
import pickle

pd.options.display.float_format = '{:,.3f}'.format

results_202105 = pd.read_pickle('results_202105.pkl')
results_202106 = pd.read_pickle('results_202106.pkl')
results_202107 = pd.read_pickle('results_202107.pkl')
results_202108 = pd.read_pickle('results_202108.pkl')
results_202109 = pd.read_pickle('results_202109.pkl')

profit_sell = 1.08 # 
loss_sell = 0.96 # 

df = pd.concat([results_202105, results_202106, results_202107, results_202108, results_202109], axis=0)
df['return'] = np.where(df['profit_sell']==1, profit_sell, np.where(df['loss_sell']==1, loss_sell, df['return_close']))
model_data = df[df['buy']==1]

print(model_data.groupby('yymmdd')['return'].mean().product())
print(model_data.groupby('yymmdd')['return'].mean().mean())

bins_dict = {}
indep = ['price_change','price_z','volume_z','price_z_out','volume_z_out','num_wins_60','num_wins_trend','price_from_5ma','num_toptail']
for var in indep:

    q1 = model_data[var].quantile(0.20)
    q2 = model_data[var].quantile(0.40)
    q3 = model_data[var].quantile(0.60)
    q4 = model_data[var].quantile(0.80)    
    
    bins_dict[var] = {}
    bins_dict[var]['q1'] = model_data[var].quantile(0.20)
    bins_dict[var]['q2'] = model_data[var].quantile(0.40)
    bins_dict[var]['q3'] = model_data[var].quantile(0.60)
    bins_dict[var]['q4'] = model_data[var].quantile(0.80)
    
bins =  [ -np.inf, bins_dict[var]['q1'], bins_dict[var]['q2'], bins_dict[var]['q3'], bins_dict[var]['q4'], np.inf]     
for var in indep:
    model_data['x_'+ var] = pd.cut(model_data[var], bins = bins, labels=range(5))
    

with open('bins_dict', 'wb') as file:  
    pickle.dump(bins_dict, file)
    
with open('bins_dict', 'rb') as file:  
    bins_dict = pickle.load(file)   

model_data['actual'] = np.where(model_data['return']>1, 1, 0)
    
cols = [c for c in model_data.columns if 'x_' in c]
y = model_data['actual']
X = model_data[cols]

classifier = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1), n_estimators=200)
classifier.fit(X, y)

model_data['yhat'] = classifier.predict_proba(X)[:,1]
model_data['select'] = np.where(model_data['yhat']  > 0.5, 1, 0)
model_data['ranks'] = pd.qcut(model_data['yhat'], q=10, duplicates='drop')
print(model_data.groupby('select')['actual'].agg(['count','mean']))
print(model_data.groupby('ranks')[['actual','return']].mean())

with open('classifier', 'wb') as file:  
    pickle.dump(classifier, file)
    
with open('classifier', 'rb') as file:  
    classifier = pickle.load(file)   
```

```python
class Circle():
    
    __pi = 3.14
    
    def __init__(self, radius):
        self.__radius = radius
#     radius = 1  #    1  .
    
    def getcircumference(self):
        result = 2*Circle.__pi*(self.__radius)
        return result
    def getarea(self):
        result = Circle.__pi*(self.__radius)**2
        return result   
    @staticmethod
    def getPI(): # self  
        return Circle.__pi
```

```python
small = Circle(10)
print(small.getarea())
print(small.getcircumference())
```
